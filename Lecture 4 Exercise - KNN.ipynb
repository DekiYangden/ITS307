# <font color = blue>Part A - Classification

## 1. Loading Data and Manipulating with Pandas

Instructions:

- Load iris dataset from sklearn.datasets module 




```python
#Your solution here
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
iris = load_iris()
bdf = pd.DataFrame(iris.data, columns = iris.feature_names)
bdf.head()
```

```python
#Your solution here
bdf['target'] = iris.target
bdf.head()

```

## 2. Train test split

Instructions:

- Split your datasets into train and test set




```python
from sklearn.model_selection import train_test_split
x = bdf.drop(['target'], axis = 1)
y = bdf.target
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 5)
```

## 3. Model Training

Instructions:

- Import KNN Classifier Algorithm from sklearn.


- Mention number of neighbours you want to choose for your model and train the model




```python
#Your solution here
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)

neigh.fit(x_train, y_train)

```

## 3. Model Evaluation

Instructions:

- Use test set to evaluate your model's performance.





```python
#Your solution here

y_pred = neigh.predict(x_train)
```

```python
neigh.score(x_train, y_train)
```

```python
neigh.score(x_test,y_test)
```

```python
neigh.predict(x_test)
```
